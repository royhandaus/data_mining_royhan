## Data Profiling

```python
# Calculate null values and percentage
null_counts = df.isnull().sum()
null_percentages = (null_counts / len(df)) * 100

# Create a DataFrame for columns with null values
null_df = pd.DataFrame({'Null Count': null_counts, 'Null Percentage': null_percentages})

# Filter for columns with null values
null_df = null_df[null_df['Null Count'] > 0]


# Format the percentage with two decimal places and % symbol
null_df['Null Percentage'] = null_df['Null Percentage'].map('{:.2f}%'.format)


null_df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create the horizontal bar plot
plt.figure(figsize=(10, 6))  # Adjust figure size as needed
sns.barplot(x='Null Count', y=null_df.index, data=null_df, orient='h', palette='rainbow')
plt.xlabel('Number of Null Values', fontsize=12)
plt.ylabel('Columns', fontsize=12)
plt.title('Null Values per Column', fontsize=14)

# Add labels to the bars
for i, v in enumerate(null_df['Null Count']):
    plt.text(v + 1, i, str(v), color='black', va='center', fontsize=10)

plt.tight_layout()
plt.show()
```

Proses ini bertujuan untuk membersihkan dan menyederhanakan tampilan angka harga properti di kolom price_in_rp. Secara default, Python akan menampilkan angka besar dalam notasi ilmiah seperti 1.5e+10, yang artinya 15 miliar. Tapi notasi ini bisa membingungkan atau tidak nyaman dibaca saat analisis. saya juga melakukan binning data dengan mengelompokkan kelas rumah berdasarkan harga property yang ditawakan seperti ini

Low: ≤ 1.5 M

Mid: 1.5 M – 4.5 M

High: > 4.5 M

```python
df['price_in_rp'] = df['price_in_rp'].round(0).astype(int)
```

```python
# Pastikan kolom harga dalam bentuk numerik
df['price_in_rp'] = pd.to_numeric(df['price_in_rp'], errors='coerce')

# Hitung batas quantile
q1 = df['price_in_rp'].quantile(0.33)
q2 = df['price_in_rp'].quantile(0.66)

print("Low ≤", q1)
print("Mid  ≤", q2)

def price_segment(price):
    if pd.isna(price):
        return 'unknown'
    elif price <= q1:
        return 'low'
    elif price <= q2:
        return 'mid'
    else:
        return 'high'

df['price_segment'] = df['price_in_rp'].apply(price_segment)
```

```python
df['property_type'].fillna('rumah', inplace=True)
```

```python
# Dictionary standar fasilitas
standard_facilities = {
    'track lari': 'jogging track',
    'kolam renang': 'kolam renang',
    'keamanan 24 jam': 'keamanan',
    'taman': 'taman',
    'masjid': 'masjid',
    'jalur telepon': 'jalur telepon',
    'one gate system': 'one gate system',
    'ac': 'ac',
    'wastafel': 'wastafel',
    'cctv': 'cctv',
    'playground': 'playground',
    'tempat jemuran': 'tempat jemuran'
}

# Fungsi pembersihan dan standarisasi
def clean_facilities(text):
    if pd.isna(text):
        return []
    raw_items = [item.strip().lower() for item in text.split(',')]
    standardized = [standard_facilities.get(item, item) for item in raw_items]
    return list(set(standardized))  # hilangkan duplikasi

# Overwrite kolom 'facilities' dengan versi bersih
df['facilities'] = df['facilities'].apply(clean_facilities)
```

```python

# Ubah ke numerik dulu agar bisa isi NaN dan ubah ke int
df['land_size_m2'] = pd.to_numeric(df['land_size_m2'], errors='coerce')
df['building_size_m2'] = pd.to_numeric(df['building_size_m2'], errors='coerce')

# Isi NaN dengan 0
df['land_size_m2'] = df['land_size_m2'].fillna(0).astype(int)
df['building_size_m2'] = df['building_size_m2'].fillna(0).astype(int)
```

```python
df['certificate'] = df['certificate'].fillna('unknown')
```

```python
# Ganti string 'nan' dengan nilai NaN asli
df['floors'] = df['floors'].replace('nan', np.nan)

# Hitung median dari nilai valid
floors = pd.to_numeric(df['floors'], errors='coerce').median()

# Isi semua NaN dengan median
df['floors'] = df['floors'].fillna(floors)
```

```python
# Ubah kolom numerik ke object agar bisa menampung string
df['building_age'] = df['building_age'].astype('object')
df['year_built'] = df['year_built'].astype('object')

# Isi semua NaN dengan 'unknown'
df['building_age'] = df['building_age'].fillna('unknown')
df['year_built'] = df['year_built'].fillna('unknown')
df['property_condition'] = df['property_condition'].fillna('unknown')
df['building_orientation'] = df['building_orientation'].fillna('unknown')
df['furnishing'] = df['furnishing'].fillna('unknown')
```

```python

def categorize_area(size):
    if pd.isna(size):
        return 'unknown'
    elif size <= 100:
        return 'small'
    elif size <= 300:
        return 'medium'
    else:
        return 'large'

df['area_category'] = df['land_size_m2'].apply(categorize_area)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.countplot(x='area_category', hue='certificate', data=df, order=['small', 'medium', 'large', 'unknown'])
plt.title('Jumlah Properti Berdasarkan Kategori Luas Tanah dan Jenis Sertifikat')
plt.xlabel('Kategori Area')
plt.ylabel('Jumlah Properti')
plt.xticks(rotation=0)
plt.legend(title='Jenis Sertifikat', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
